```python
from sklearn.preprocessing import LabelEncoder
l=['Car_Name','Fuel_Type','Seller_Type','Transmission']
le=LabelEncoder()

for i in l:
  df[i]=le.fit_transform(df[i])
```

```python
x=df.drop(columns=['Selling_Price'])
y=df['Selling_Price']
```

```python
x=pd.get_dummies(x)
df.shape
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
se=StandardScaler()
```

```python
x_train=se.fit_transform(x_train)
x_test=se.transform(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error
def evaluate_model(model,x_train,x_test,y_train,y_test):
  train_pred=model.predict(x_train)
  test_pred=model.predict(x_test)
  print(f"Train r2_score for {model} is {r2_score(y_train,train_pred)}")
  print(f"Test r2_score for {model} is {r2_score(y_test,test_pred)}")
  print(f"Train MAE for {model} is {mean_absolute_error(y_train,train_pred)}")
  print(f"Test MAE for {model} is {mean_absolute_error(y_test,test_pred)}")
  print(f"Train RMSE for {model} is {mean_absolute_error(y_train,train_pred)}")
  print(f"Test RMSE for {model} is {mean_absolute_error(y_test,test_pred)}")
```

```python
from sklearn.linear_model import LinearRegression,Ridge
LinearRegression_model=LinearRegression()
LinearRegression_model.fit(x_train,y_train)
```

```python
evaluate_model(LinearRegression_model,x_train,x_test,y_train,y_test)
```

```python
y_pred=LinearRegression_model.predict(x_test)
y_pred_train=LinearRegression_model.predict(x_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
```

```python
ridge_param={
    "alpha":[1,2,10,50,100]
}
```

```python
ridge_search=GridSearchCV(ridge,ridge_param,cv=5)
```

```python
ridge_search.fit(x_train,y_train)
```

```python
best_ridge=ridge_search.best_estimator_
```

```python
evaluate_model(best_ridge,x_train,x_test,y_train,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_moedel=KNeighborsRegressor()
```

```python
knn_param={
    "n_neighbors":[1,3,5,7,9]
}
```

```python
knn_search=GridSearchCV(knn_moedel,knn_param,cv=5)
```

```python
knn_search.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model=DecisionTreeRegressor()
```

```python
dt_param={
    'max_depth':[3,5,7,9,11],
    'min_samples_split':[2,5,10,15],
    'min_samples_leaf':[1,2,4,6]
}
```

```python
tree_search=GridSearchCV(dt_model,dt_param,cv=5)
```

```python
tree_search.fit(x_train,y_train)
```

```python
best_tree=tree_search.best_estimator_
best_tree
```

```python
evaluate_model(best_tree,x_test,x_train,y_test,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor()
```

```python
rf_param={
    'n_estimators':[50,100,150],
    'max_depth':[5,10,15],
    'max_features':['log2','sqrt']
}
```

```python
rf_search=GridSearchCV(rf_model,rf_param,cv=5)
```

```python
rf_search.fit(x_train,y_train)
```

```python
best_rf=rf_search.best_estimator_
```

```python
evaluate_model(best_rf,x_train,x_test,y_train,y_test)
```
